## Loaders

### Docx loader

````python
from langchain_community.document_loaders import Docx2txtLoader

loader = Docx2txtLoader("documents/Abreviaciones 1.docx")
data = loader.load()
````

````python
data[0].dict()
````

````python
with open('page_content.txt', 'w') as file:
    file.write(data[0].page_content)
````

## Text Splitters

Tenemos el problema de que si hay una estrucutra como la siguiente:

```txt
RAE
Real Academia Española
```

y el chunk se corta entre RAE y la explicacion de ese acronimo se pierde ese contexto. Se soluciona con un pequeño overlap entre los chunks.

Posible mejora, splitter semantico:
https://python.langchain.com/docs/modules/data_connection/document_transformers/semantic-chunker/

````python
text = data[0].page_content
text
````

````python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", "  ", '.'],
    chunk_size=500,
    chunk_overlap=50
)
````

````python
# from langchain.text_splitter import CharacterTextSplitter

# splitter = CharacterTextSplitter(
#     separator = "\n",
#     chunk_size=200,
#     chunk_overlap=50
# )
````

````python
chunks = splitter.split_text(text=text)
````

````python
for chunk in chunks:
    print(len(chunk))
    print(chunk)
    print("-"*50)
````
